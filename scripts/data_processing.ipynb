```python
import shutil
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

## Настройки

```python
DATA_PATH = '/content/drive/MyDrive/Colab Notebooks/ОБУЧЕНИЕ/netology.ru/2. Менеджмент data-проектов/4. Сохранение результатов эксперимента/Задание/data/'
```

## Копируем файл .csv

```python
shutil.copyfile('/content/winequalityN.csv', DATA_PATH + 'external/' + 'winequalityN.csv')
```

## Обработка данных

```python
data = pd.read_csv(DATA_PATH + 'external/winequalityN.csv')
```

```python
# Заполняем средними
data_new = data.fillna(data.mean())
```

```python
# Переводим категориальные данные в числовые
data_dum = pd.get_dummies(data_new)

# В numpy
data_result = data_dum.to_numpy()
```

## Сохраняем numpy

```python
np.save(DATA_PATH + 'processed/data.npy', data_result)
```
